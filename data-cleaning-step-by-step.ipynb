```python
def identify_column_types(df):
   
    numerical_columns = []
    categorical_columns = []

    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            numerical_columns.append(column)
        elif pd.api.types.is_object_dtype(df[column]):
            categorical_columns.append(column)

    return numerical_columns, categorical_columns
```

```python
numerical_cols, categorical_cols = identify_column_types(df)
print("Numerical Columns:", numerical_cols)
print("Categorical Columns:", categorical_cols)
```

```python
import pandas as pd

def handle_missing_values(df, numerical_columns, categorical_columns, numerical_strategy='mean', categorical_strategy='mode'):
 
    for column in numerical_columns:
        if df[column].isna().any():
            if numerical_strategy == 'mean':
                df[column].fillna(df[column].mean(), inplace=True)
            elif numerical_strategy == 'median':
                df[column].fillna(df[column].median(), inplace=True)
            else:
                pass


    for column in categorical_columns:
        if df[column].isnull().any():
            if categorical_strategy == 'mode':
                df[column].fillna(df[column].mode()[0], inplace=True)
            else:
                pass

    return df

numerical_cols, categorical_cols = identify_column_types(df)

print(numerical_cols)
print(categorical_cols)

print("Original DataFrame:")
df.tail()

df = handle_missing_values(df, numerical_cols, categorical_cols, numerical_strategy='mean', categorical_strategy='mode')

print("\nDataFrame after handling missing values:")
df.tail()
```
